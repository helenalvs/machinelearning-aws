```python
import boto3
import json
from matplotlib import pyplot as plt

from PIL import Image, ImageDraw, ImageColor, ImageOps

client=boto3.client('rekognition')
    
def recognize_celebrities(photo):

    with open(photo, 'rb') as image:
        response = client.recognize_celebrities(Image={'Bytes': image.read()})

    print('Detectando rostos em ' + photo)    
    name = []
    for celebrity in response['CelebrityFaces']:
        name.append(celebrity['Name'])
        #print ('Nome: ' + celebrity['Name'])
        #name = celebrity['Name']
        #print ('Id: ' + celebrity['Id'])
        #print ('Posição:')
        #print ('   Esquerda: ' + '{:.2f}'.format(celebrity['Face']['BoundingBox']['Height']))
        #print ('   Superior: ' + '{:.2f}'.format(celebrity['Face']['BoundingBox']['Top']))
        #print ('Info')
        #for url in celebrity['Urls']:
        #    print ('   ' + url)
    return len(response['CelebrityFaces']), name

photo_list = ['test1.png', 'viola.jpg', 'emma.jpg', 'How-i-met-your-mother.jpg']
total_celebs = 0
celebs_name = []
for i in range(len(photo_list)):
    celeb_count, celeb_name=recognize_celebrities(photo_list[i])
    total_celebs += celeb_count
    for name in celeb_name:
        celebs_name.append(name)
print("Celebridades detectadas: " + str(total_celebs))
print("Nome das celebridades encontradas:")
count = 0
for name in celebs_name:
    count +=1
    print(count, ". ", name, sep='')
```

```python
imgs = []
for photo in photo_list:
    img = Image.open(photo)
    imgWidth, imgHeight = img.size

    with open(photo, 'rb') as image:
            response = client.recognize_celebrities(Image={'Bytes': image.read()})
    draw = ImageDraw.Draw(img)
    for celebrity in response['CelebrityFaces']:
        box = celebrity['Face']['BoundingBox']
        left = imgWidth * box['Left']
        top = imgHeight * box['Top']
        width = imgWidth * box['Width']
        height = imgHeight * box['Height']

        points = ((left,top),(left+width,top),(left+width,top+height),(left,top+height),(left,top))

        draw.line(points,fill='#00d400', width=7)
    imgs.append(img)

print('Caixa delimitadora para identificar o rosto da(s) celebridade(s)')
for img in imgs:
    imgplot = plt.imshow(img)
    plt.show()
    
```
